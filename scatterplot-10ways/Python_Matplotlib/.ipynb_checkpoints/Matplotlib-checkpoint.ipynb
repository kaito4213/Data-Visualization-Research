```python
import matplotlib.pyplot as plt
from matplotlib.mlab import csv2rec
import pandas as pd
import csv
%matplotlib inline
```

```python
f = open('../cars-sample.csv')
cars = csv.reader(f)
header = next(cars)
weight = []
mpg = []
area = []
color = []
for row in cars:
    if (row[3] == 'NA' or row[7] == 'NA'):
        row[3] = 0
        row[7] = 0
    mpg.append(float(row[3]))
    weight.append(float(row[7]))
    area.append(3.14*(float(row[7])*0.001)**2) 
    #  ! TODO
    if row[2]=='ford':
        row[2] = 0
    else:
        if row[2]=='bmw':
            row[2] = 1
        else:
            if row[2]=='mercedes':
                row[2]=2
            else:
                if row[2]=='honda':
                    row[2]=3
                else:
                    row[2]=4
        
    color.append(row[2])

print(color)
plt.scatter(weight, mpg, s=area, c=color, alpha = 0.5)
plt.axis([1000, 5000,8, 50,])# todo
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.show()
```
