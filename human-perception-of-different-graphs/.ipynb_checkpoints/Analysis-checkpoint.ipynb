```python
import pandas as pd
import matplotlib as plt
import glob
%matplotlib inline
```

```python
filename = glob.glob("data/*.csv")
list = []
for file in filename:
    df = pd.read_csv(file,index_col = None, header = 0)
    list.append(df)
data = pd.concat(list)
data.head(n = 5)
```

```python
ErrorofType = data[['plottype','accuracy']]
bp = ErrorofType.boxplot(by = 'plottype')
```

```python
data.loc[data['act_percent'] <= 33,'No.'] = '<33'
data.loc[((data['act_percent'] >= 33)&(data['act_percent'] <= 66)),'No.'] = '33~66'
data.loc[data['act_percent'] >= 66,'No.'] = '>66'
meanofType = data['accuracy'].groupby(data['No.']).mean()
meanofType = pd.DataFrame(meanofType)
meanPlot = meanofType.plot(kind='bar')
```

```python
import seaborn as sns
import numpy as np
data_list = []
for i in range(10):
    boot_index = np.random.choice(range(600),size = 600, replace = True)
    new_data = data.ix[boot_index]
    data_list.append(new_data)
data_boot = pd.concat(data_list, axis = 0)
ax = sns.boxplot(x = 'plottype', y = 'accuracy', data = data_boot, whis = np.inf)
```
